```python
def sort_impact(x):
    if x <= 10:
        return 'impacted States'
    elif x <= 20:
        return 'hard impacted States'
    elif x <= 30:
        return 'harder impacted States'
    elif x <= 40:
        return 'hardest impacted States'
    return x
```

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Define lockdown date
lockdown_date = pd.to_datetime('24-03-2020', format='%d-%m-%Y')

def sort_by_lockdown_impact(data):
    # Filter data after lockdown
    after_lockdown = data[data['Date'] >= lockdown_date]

    # Calculate average unemployment rate after lockdown for each state
    average_unemployment_after_lockdown = after_lockdown.groupby('State')['Estimated_Unemployment_Rate'].mean()

    # Calculate percentage change in unemployment rate compared to before lockdown
    percentage_change = ((average_unemployment_after_lockdown - data[data['Date'] < lockdown_date].groupby('State')['Estimated_Unemployment_Rate'].mean()) / data[data['Date'] < lockdown_date].groupby('State')['Estimated_Unemployment_Rate'].mean()) * 100

    # Sort states based on percentage change
    sorted_states = percentage_change.sort_values(ascending=False)

    return sorted_states

# Test the function
sorted_states = sort_by_lockdown_impact(df)
print("Sorted States based on Impact of COVID-19 Lockdown:")
print(sorted_states)
```
